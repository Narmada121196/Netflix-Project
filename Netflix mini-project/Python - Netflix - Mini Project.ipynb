```python
#The fillna() method replaces the NULL values with a specified value. 
df['director'].fillna('No Director', inplace = True)
```

```python
df.isna().sum()
```

#### Task 1: Plot a graph showing number of contents on Netflix based on Rating.

```python
df['rating'].value_counts()
```

```python
df['rating'].value_counts().plot(kind='bar',color = 'green',edgecolor = 'Red')
plt.xlabel('Contents in Netflix')
plt.ylabel('Ratings')
plt.title('Number of contents on Netflix based on Rating')
```

### Conclusion:

Based on ratings, TV-MA (Mature Audience Only) has more contents.
So, Adults has more contents in Netflix to look than children.
Netflix has least contents on the rating NC-17.

#### Task 2: Plot a graph showing the number of TV Shows and Movies across different type of Rating.


```python
df.columns
```

```python
df['rating'].groupby(df['type']).value_counts().plot(kind = 'bar',color = 'black',edgecolor = 'red')
plt.xlabel('TV Shows vs Movies in Netflix')
plt.ylabel('Ratings')
plt.title('Number of TV Shows and Movies across different type of Rating')
```

```python
df['type'].groupby(df['rating']).value_counts().plot(kind = 'bar',color = 'black',edgecolor = 'red')
plt.xlabel('TV Shows vs Movies in Netflix')
plt.ylabel('Ratings')
plt.title('Number of TV Shows and Movies across different type of Rating')
```

### Conclusion:
Thus, to specify the contents based on ratings in Netflix, I conclude that overall, "Movies with ratings TV-MA" has more 
contents than all other ratings as well as the TV-Shows.
In TV-Shows, we see that "TV-14" which children under 14 are not suitable for contents are more in Netflix.

"NC-17" in Movies and "G" in TV-Show has the least contents in Netflix.

#### Task 3: Compare India, US, Japan, Spain, UK & Canada to check which country produces highest TV Show and Movies.

```python
df.columns
```

```python
country = ['India','United States','Japan','Spain','United Kingdom','Canada']
ctry = df.loc[df['country'].isin(country)]
ctry.head()
```

```python
df['type'].groupby(ctry['country']).value_counts().plot(kind = 'bar',color = 'white',edgecolor = 'red',lw = 2)
plt.xlabel('TV Shows vs Movies Country Specific')
plt.ylabel('Ratings')
plt.title('Number of TV Shows and Movies across different type of Countries')
```

```python
ctry['country'].groupby(df['type']).value_counts().plot(kind = 'bar',color = 'cyan',edgecolor = 'red',lw=1.2)
plt.xlabel('TV Shows vs Movies Country Specific')
plt.ylabel('Ratings')
plt.title('Number of TV Shows and Movies across different type of Countries')
```

### Conclusion:
I conclude that Netflix in United States tops the Movie Contents followed by India.
And in TV Shows also USA tops the number the contents followed by United Kingdom.

Thus, movies contents in Japan and TV-show contents in Spain has fewer contents in Netflix.

#### Task 4: Plot a graph showing the number of TV Shows and Movies added to Netflix every year.

```python
df.columns
```

```python
Date = pd.to_datetime(df['date_added'])
year = Date.dt.yearx
```

```python
df['year'] = Date.dt.year
```

```python
df.head(2)
```

```python
df['year'].groupby(df['type']).value_counts()
```

```python
#showing the number of TV Shows and Movies added to Netflix every year.

df['type'].groupby(df['year']).value_counts().plot(kind='bar',color = 'orange',edgecolor = 'black',lw = 1.0,linestyle = '-')
plt.xlabel('TV Shows vs Movies year_wise')
plt.ylabel('Ratings')
plt.title('Number of TV Shows and Movies across different years')
```

```python
df['year'].groupby(df['type']).value_counts().plot(kind='bar',color = 'blue',edgecolor = 'black',lw = 1.0,linestyle = '--')
plt.xlabel('TV Shows vs Movies year_wise')
plt.ylabel('Ratings')
plt.title('Number of TV Shows and Movies across different years')
```

```python
df['year'].groupby(df['type']).value_counts()
```

### Conclusion:
We can conclude that contents are gradually added each year at the rate of 0.25% to 0.50%. There is no decrease in between the years rather it is a gradual increase happening over the years in both TV Shows as well as Movies.

#### Task 5: In 2019, Find out if Netflix has added more TV Shows or Movies to their platform and represent the same visually.

```python
year_2019 = df[df['year'] == 2019]
```

```python
year_2019['type'].value_counts().plot(kind = 'bar', color = 'cyan',edgecolor = 'orange', width = 0.3, lw = 3.0,linestyle = '--')
plt.xlabel('Movies vs TV Shows')
plt.title('Movies vs TV Shows in Netflix')
```

### Conclusion:
It is obvious that in 2019, Netflix has more contents on "movies" than "TV-Shows".
